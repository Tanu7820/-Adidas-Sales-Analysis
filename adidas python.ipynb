```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\TANUSHREE\Downloads\Adidas US Sales Datasets.xlsx")
```

```python
df.info
```

```python
df.columns = df.iloc[2]   # row 2 ko header bana do
df = df[3:]               # top rows remove
df = df.reset_index(drop=True)
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
cols = ['Price per Unit', 'Units Sold', 'Total Sales', 
        'Operating Profit', 'Operating Margin']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
```

```python
df.isnull()
```

```python
df.columns
```

```python
df.columns = df.columns.fillna("Remove")
```

```python
df = df.drop(columns=["Remove"])
```

```python
df.columns
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Total Sales'], bins=20)
plt.title("Distribution of Total Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
region_sales = df.groupby('Region')['Total Sales'].sum()

plt.bar(region_sales.index, region_sales.values)
plt.title("Region vs Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Month'] = df['Invoice Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Total Sales'].sum()

plt.plot(monthly_sales.index.astype(str), monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)

plt.figure()
plt.bar(top_products.index[:5], top_products.values[:5])
plt.title("Top 5 Products by Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
profit_region = df.groupby('Region')['Operating Profit'].sum()

plt.figure()
plt.bar(profit_region.index, profit_region.values)
plt.title("Profit by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
📌 Final Insights
🏆 West/Northeast region shows the highest sales and profit contribution, indicating strong market demand in these areas.
👟 Men’s Street Footwear and Athletic Footwear are the top-performing product categories in terms of revenue generation.
📈 Sales trend shows fluctuations over time with peak months indicating seasonal demand patterns.
🏬 In-store sales method dominates overall transactions, showing offline channel is still strongest for this business.
💰 There is a positive relationship between Units Sold and Total Sales, meaning higher volume directly increases revenue.
📊 Profit distribution varies by region, showing that high sales does not always mean high profit efficiency.
```
